```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import spearmanr
```

# Original data

```python

earwig_density = np.array([0.3, 5.2, 2.5, 25.6, 8.1, 0.3, 4.7, 3.3, 7.8, 20, 31, 25, 43.3, 33.9, 33.9, 32.7, 33.8, 12.7, 57, 52.5, 64, 70.4])
pro_male_forceps = np.array([0.04, 0.02, 0.07, 0.06, 0.13, 0.15, 0.2, 0.24, 0.25, 0.19, 0.22, 0.32, 0.3, 0.44, 0.52, 0.55, 0.62, 0.66, 0.46, 0.38, 0.38, 0.46])
```

# Histogram

```python
# Histogram for earwig density
plt.subplot(1, 2, 1)
plt.hist(earwig_density, bins=10, edgecolor='black')
plt.title('Histogram of Earwig Density')
plt.xlabel('Earwig Density')
plt.ylabel('Frequency')

# Histogram for proportion of males with forceps
plt.subplot(1, 2, 2)
plt.hist(pro_male_forceps, bins=10, edgecolor='black')
plt.title('Histogram of Proportion of Males with Forceps')
plt.xlabel('Proportion of Males with Forceps')
plt.ylabel('Frequency')


plt.tight_layout()
plt.show()
```

# Scatter Plot

```python
plt.scatter(earwig_density, pro_male_forceps)

plt.xlabel("Earwig Density", fontsize = 14)
plt.ylabel("Proportion of Males with Forceps", fontsize = 14)

plt.show()
```

# Log Transformation

```python

# Apply log transformation to the earwig density
log_earwig_density = np.log(earwig_density + 1e-6)  # Adding a small constant to avoid log(0)

# Plotting the transformed data
plt.scatter(log_earwig_density, pro_male_forceps)

plt.xlabel("Log-transformed Earwig Density", fontsize = 14)
plt.ylabel("Proportion of Males with Forceps", fontsize = 14)

plt.show()
```

```python
# Histogram for log-transformed earwig density
plt.subplot(1, 2, 1)
plt.hist(log_earwig_density, bins=10, edgecolor='black')
plt.title('Histogram of Log-transformed Earwig Density')
plt.xlabel('Log-transformed Earwig Density')
plt.ylabel('Frequency')

# Histogram for proportion of males with forceps
plt.subplot(1, 2, 2)
plt.hist(pro_male_forceps, bins=10, edgecolor='black')
plt.title('Histogram of Proportion of Males with Forceps')
plt.xlabel('Proportion of Males with Forceps')
plt.ylabel('Frequency')


plt.tight_layout()
plt.show()
```

# Spearman's Rank Correlation and t statistic

```python
correlation, p_value = spearmanr(log_earwig_density, pro_male_forceps)
n = len(earwig_density)

test_statistic = (correlation * np.sqrt(n - 2)) / np.sqrt(1 - correlation**2)


print(f"Spearman's rank correlation: {correlation:.3f}")
print(f"P-value: {p_value:.3f}")
print(f"Test Statistic (t): {test_statistic:.3f}")
```
